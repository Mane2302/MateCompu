**Tipos de texto**

```python
class texto(Scene):
    def construct(self):
        textoText = Text("Hola Mundo en Text").shift(UP)
        textoText.set_color(BLUE) #color
        textoMathTex = MathTex(r"e^{\pi i}-1=0 \: en MathTex")
        textoTex = Tex("Hola Mundo: $f(x)= x^2$").shift(DOWN)
        #texto.scale(2)

        #texto.set_color_by_gradient([RED,BLUE_A])
        self.play(Write(textoText),
                  Write(textoMathTex),
                  Write(textoTex),
                  run_time= 3)
        self.wait(2)
```

**Colorear Texto**

```python
%%manim -v WARNING -qm ColoreadoTexto
class ColoreadoTexto(Scene):
    def construct(self):
        texto = Text("Texto u objeto")
        self.add(texto)
        self.wait(0.5)
        for texto in texto:
            self.play(LaggedStart(*[letra.animate.set_color(YELLOW) for letra in texto],
                run_time = 0.12
            ))
        self.wait(0.5)
```

**Llaves**

```python
class Llave(Scene):
    def construct(self):
        texto=MathTex(
            "\\frac{d}{dx}f(x)g(x)=","f(x)\\frac{d}{dx}g(x)","+",
            "g(x)\\frac{d}{dx}f(x)"
        )
        self.play(Write(texto))
        #asignamos las llaves y la  direccion
        llave_s = Brace(texto[1], UP, buff = SMALL_BUFF)
        llave_i = Brace(texto[3], DOWN, buff = SMALL_BUFF)
        #agregamos texto a  las llaves
        t_s = llave_s.get_text("$g'f$").set_color(YELLOW)
        t_i = llave_i.get_text("$f'g$")
        self.play(
            GrowFromCenter(llave_s),
            GrowFromCenter(llave_i),
            FadeIn(t_s),
            FadeIn(t_i)
            )
        self.wait()
```

## Primer mini proyecto

```python
%%manim -v WARNING -qm Proyecto1
class Proyecto1(Scene):
    def construct(self):
        #self.add(NumberPlane())
        titulo = Title("Ecuación Cuadratica").scale(1.5)
        texto1 = Text("Son de la forma:").to_corner(UL).shift(DOWN).scale(0.8)
        texto2 = Text("Cuyas soluciones están dadas por:").scale(0.8).to_corner(UL).shift(3*DOWN + RIGHT*0.5)
        eq1 = MathTex(r"ax^2+bx+c=0").shift(UP*1.5)

        eq2 = MathTex(r" x_{1,2}=\frac{-b\pm \sqrt{b^2-4ac}}{2a}").shift(DOWN*2)

        grupo = VGroup(titulo, texto1, texto2, eq1)

        self.play(Write(titulo),  run_time = 2)
        self.wait()
        self.play(Write(texto1))
        self.wait()
        self.play(Write(eq1))
        self.wait()
        self.play(Write(texto2))
        self.wait()
        self.play(Write(eq2))
        self.wait()
        self.play(FadeOut(grupo))
        self.wait(2)
        self.play(eq2.animate.shift(2*UP))
        self.wait(2)
        self.play(eq2.animate.scale(2))
        self.wait(2)
        self.play(eq2.animate.set_color(RED))
        self.wait(2)
```

## Animando figuras

```python
# Definimos una clase que hereda de Scene para crear una animación básica
class BasicAnimations(Scene):
    # Método principal construct donde se definen los objetos y animaciones
    def construct(self):
        # Creamos  las figuras: cuadrado, triángulo y círculo
        # Cada forma se escala para ser un 50% más grande
        cuadrado = Square().scale(1.5)
        triangulo = Triangle().scale(1.5)
        circulo = Circle().scale(1.5)

        # Animamos cada figura de manera simultánea
        self.play(
            Create(cuadrado, run_time = 2),       # Se crea el cuadrado con una duración de 2 segundos
            FadeIn(triangulo, run_time = 2),      # El triángulo aparece gradualmente en 2 segundos
            GrowFromCenter(circulo, run_time = 2) # El círculo crece desde el centro en 2 segundos
        )
        # Pausamos la escena durante 2 segundos antes de terminar
        self.wait(2)
```

```python
class Fondo(Scene):
    def construct(self):
        # Establece el color de fondo de la cámara usando un valor hexadecimal
        self.camera.background_color="#F0E68C"

        # Aplicamos colores a cada figura
        cuadrado = Square(color=BLACK).scale(1.5)
        triangulo = Triangle(color=RED).scale(1.5)
        circulo = Circle(color=PINK).scale(1.5)

        self.play(
            Create(cuadrado, run_time = 2),
            FadeIn(triangulo, run_time = 2),
            GrowFromCenter(circulo, run_time = 2)
            )
```

```python
class Punto(Scene):
    def construct(self):
        plano =  NumberPlane().add_coordinates()
        self.add(plano)

        punto1 = Dot(point=[2, 3, 0],
                    radius=0.1,
                    #stroke_width=2
                    )

        punto2 = Dot(np.array([-2, 1, 0]), color= RED)

        self.play(Create(punto1))
        self.wait()
        self.play(Create(punto2))
        self.wait()
```

```python
class Poligono(Scene):
    def construct(self):
        plano =  NumberPlane().add_coordinates()
        self.add(plano)

        #creamos los puntos vertices
        p1 = np.array([-3,2,0])
        p2 = np.array([3,2,0])
        p3 = np.array([5,-2,0])
        p4 = np.array([-5,-3,0])

        punto1 = Dot(p1)
        punto2 = Dot(p2)
        punto3 = Dot(p3)
        punto4 = Dot(p4)

        poligono = Polygon(p1, p2, p3, p4).set_fill(color=PURPLE, opacity=0.5)

        self.play(Create(punto1),
                  Create(punto2),
                  Create(punto3),
                  Create(punto4))
        self.wait()


        self.play(Create(poligono))
        self.wait()
```
